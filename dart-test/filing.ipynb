```python
import dart_fss as dart
import pandas as pd
import os
from dotenv import load_dotenv

# set up api keys from env
load_dotenv()
api_key = os.getenv("DART_API_KEY")
dart.set_api_key(api_key=api_key)

# get corp list
corp_list = dart.get_corp_list()

# find samsung's info by name
# https://dart-fss.readthedocs.io/en/latest/dart_corp.html
samsung = corp_list.find_by_corp_name('삼성전자', exactly=True)[0]

samsung.corp_code
```

```python
# search filings from corp class
samsung_filing = samsung.search_filings(bgn_de='20220101', end_de='20220301')
samsung_filing
```

```python
# directly search filings
samsung_filing = dart.filings.search(corp_code=samsung.corp_code, bgn_de='20220101', end_de='20220301')
samsung_filing
```
