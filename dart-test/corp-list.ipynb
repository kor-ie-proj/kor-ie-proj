```python
import dart_fss as dart
import pandas as pd
import os
from dotenv import load_dotenv

# set up api keys from env
load_dotenv()
api_key = os.getenv("DART_API_KEY")
dart.set_api_key(api_key=api_key)

# fetch corp codes data and create dataframe
all_corps = dart.api.filings.get_corp_code()
all_corps_df = pd.DataFrame(all_corps)

all_corps_df
```

```python
# save it to excel  (if no directory exists, create it)
if not os.path.exists('data'):
    os.makedirs('data')
all_corps_df.to_excel(f"data/corp_list.xlsx", index=False)
```
